```python
columns_to_drop = ['Patient ID','Patient Name','Residence Type','Alcohol Intake','Physical Activity','Cholesterol Levels','Symptoms','Marital Status','Dietary Habits','Work Type','Blood Pressure Levels','Stress Levels','Stroke History','Gender','Family History of Stroke']
```

```python
stroke_df = stroke_df.drop(columns_to_drop, axis=1)
```

```python
stroke_df
```

```python
age_categories = {range(18,50):0,range(51,105):1}

def category_range(age):
    for age_range,category in age_categories.items():
        if age in age_range:
            return category
    return -1
    
stroke_df['Age'] = stroke_df['Age'].apply(category_range)
```
